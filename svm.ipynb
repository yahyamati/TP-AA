```python
from sklearn import datasets
from sklearn.svm import SVC
import matplotlib.pyplot as plt
```

```python
# Charger un jeu de données
data = datasets.make_blobs(n_samples=100, centers=2, random_state=6)
X, y = data

```

```python
# Entraîner un modèle SVM linéaire
model = SVC(kernel='linear')
model.fit(X, y)
```

```python
# Extraire les vecteurs supports
support_vectors = model.support_vectors_

# Afficher les données et l'hyperplan
plt.scatter(X[:, 0], X[:, 1], c=y, s=30, cmap="Paired")  # Corrected colormap usage
plt.scatter(
    support_vectors[:, 0],
    support_vectors[:, 1],
    s=100,
    facecolors="none",
    edgecolors="k",
    label="Vecteurs Supports",
)
plt.legend()
plt.show()
```
